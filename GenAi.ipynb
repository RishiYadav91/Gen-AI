```python
import nltk
import json
import string

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
text ="""
my name is rishi yadav and i am pursuing bachelors in computer science
"""
corpus = text.lower()
print(corpus)
```

```python
nltk.download('punkt_tab')
tokens = word_tokenize(corpus)
print(tokens)
```

```python
tokens = [word for word  in tokens if word not in string.punctuation]
print(tokens)
```

```python
stop_words = set(stopwords.words('english'))

clean_tokens = [word for word in tokens if word not in stop_words]
print(clean_tokens)
```

```python
vocabulary = sorted(set(clean_tokens))
print(vocabulary)
```

```python
vocab_dict = {"vocabulary": vocabulary}

with open("vocabulary.json", "w") as file:
    json.dump(vocab_dict, file,indent=4)

print("Vocabulary saved to vocabulary.json")
```
